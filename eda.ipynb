# Exploratory Data Analysis

**Import Libraries**

```python
import numpy as np  # For math operations
import seaborn as sns  # For making plots
import matplotlib.pyplot as plt  # For creating charts
import pandas as pd  # For working with data
```

**Read Excel Data**

```python
df = pd.read_excel('1 - Project Data.xlsx')
```

## Explore the Data / Quality Check

**Display the Data**

```python
pd.set_option('display.max_columns', None) ## show all cols, scrollable as opposed to collapsed view
df.head()
```

**Check the Shape, num of rows and cols**

```python
df.shape
```

**Check the Data Types**

```python
df.dtypes
```

**Check for Nulls**

```python
df.isnull().sum()
```

**Check Unique Values in Cols**

```python
categorical_columns = df.select_dtypes(include='object').columns

for col in categorical_columns:
    print(f"\nUnique values in '{col}':")
    print(df[col].unique())
```

```python
categorical_columns = df.select_dtypes(include='object').columns

for col in categorical_columns:
    print(f"\nUnique values in '{col}':")
    print(df[col].value_counts())
```

*Target == Churn rate (Churn Label or Churn Value)*

## Visualise the Data

**Check the Churn Values against each other**

```python
sns.countplot(x='Churn Label', data=df)
plt.title('Churn Distribution')
plt.show()
```

**Churn Rate vs Categorical Values**

```python
# Churn by gender
sns.countplot(x='Gender', hue='Churn Label', data=df)
plt.title('Churn by Gender')
plt.show()

# Senior Citizen vs Churn
sns.countplot(x='Senior Citizen', hue='Churn Value', data=df)
plt.title('Churn by Senior Citizen Status')
plt.show()

# Churn by contract
sns.countplot(x='Contract', hue='Churn Value', data=df)
plt.title('Churn by Contract type')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() # Adjust layout
plt.show()

# Churn by Internet Service
sns.countplot(x='Internet Service', hue='Churn Value', data=df)
plt.title('Churn by Internet Service')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() # Adjust layout
plt.show()

```

**Count of Churn Reasons**

```python
order = df['Churn Reason'].value_counts().index
plt.figure(figsize=(12, 12))
sns.countplot(x='Churn Reason', hue='Churn Reason', data=df, order=order)
plt.title('Churn by Reason')
plt.xticks(rotation=45, ha='right')
plt.legend(loc = 'upper left', bbox_to_anchor=(1, 1))
plt.show()
```

**Churn by Tenure and Monthly Charges**

```python
# Tenure vs Churn
sns.histplot(data=df, x='Tenure Months', hue='Churn Value', kde=True)
plt.title('Tenure Distribution by Churn')
plt.show()

# Monthly Charges vs Churn
sns.boxplot(x='Churn Value', y='Monthly Charges', data=df, palette=['#749072'])
plt.title('Monthly Charges by Churn')
plt.ylabel('Monthly Charges ($)')
plt.xlabel('Customers')
plt.xticks(ticks=[0, 1], labels=['Customers who have not churned', 'Customers that have churned'])
plt.show()
```

**Monthly Charges vs Total Charges**

```python
df['Total Charges'] = pd.to_numeric(df['Total Charges'], errors='coerce')
df_clean = df.dropna(subset=['Total Charges', 'Monthly Charges'])

sns.scatterplot(data=df_clean, x='Monthly Charges', y='Total Charges', hue='Churn Label')
plt.title('Monthly vs. Total Charges by Churn Status')
plt.show()
```

**Correlation Heat Map**

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(7,5))
sns.heatmap(corr, annot=True, cmap='rocket', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

**Churn by Multiple Services**

```python
services = ['Phone Service', 'Online Security', 'Online Backup', 'Tech Support', 'Streaming TV', 'Streaming Movies']

for service in services:
    sns.countplot(x=service, hue='Churn Label', data=df)
    plt.title(f'Churn by {service}')
    plt.show()
```

**Explore Feature Correlation**

```python
df_num = pd.get_dummies(df, columns = ['Gender','Senior Citizen','Partner','Dependents','Phone Service','Multiple Lines','Internet Service','Online Security','Online Backup','Device Protection','Tech Support', 'Streaming TV','Streaming Movies','Contract','Paperless Billing','Payment Method'], dtype=int)

df_num.head()
```

```python
# Compute correlation matrix
correlation_matrix = df_num.corr(numeric_only=True)

# Extract correlation values with target variable 'Life_expectancy'
target_corr = correlation_matrix[['Churn Value']].sort_values(by='Churn Value', ascending=False)

# Plot heatmap
plt.figure(figsize=(8, 10))
sns.heatmap(target_corr, cmap="CMRmap", annot=True, fmt=".2f", linewidths=0.5)

# Title and display
plt.title("Feature Correlation with Churn Value")
plt.show()
```
